#Tutorial for using the model demo.

To use the model, you'll need to load both the model itself and a dataset of images. This brief tutorial will guide you through the process of utilizing the model, using the files available in the GitHub folder provided.

````python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import os
import keras
from keras.preprocessing.image import load_img
from tensorflow.keras.utils import to_categorical
````

##Load Model

First thing we need to do is to load the model using the next function.
```
tf.keras.saving.load_model(
    filepath, custom_objects=None, compile=True, safe_mode=True, **kwargs
)
```
Ensure that you have all the necessary files to use the model, and make sure to adjust the 'filepath' variables according to the location of the downloaded or stored files. In this example, the files and paths used are specific to the implementation environment of the Google Colab notebook being employed.  


````python
#Edit this variable
MODEL_DIR = '/content/AiModel.keras'
````

````python
model = tf.keras.saving.load_model(MODEL_DIR)
````

##Load the data

The model was trained on the [Face expression recognition dataset](https://www.kaggle.com/datasets/jonathanoheix/face-expression-recognition-dataset/) sourced from Kaggle. This dataset exclusively consists of 48x48 grayscale images. As the deployed model is designed to work with images sharing these specific attributes, any image that deviates from these characteristics must undergo preprocessing prior to use. In the code below, we will load the images from the validation folder within the same dataset.

````python
#Edit this variable
VALIDATION_DIR = '/content/images/validation'
````

````python
#We define the labels used
label = ['angry','disgust','fear','happy','neutral','sad','surprise']
#We define the features of the image to be loaded.
def ef(image):
    img = load_img(image,color_mode = "grayscale" )
    feature = np.array(img)
    feature = feature.reshape(1,48,48,1)
    return feature
````

##Using the Model

With both the model and the images loaded, we are now ready to utilize the dataset with the 'predict' function. In the following section, we will showcase a few examples using individual images

````python
image1 = VALIDATION_DIR + '/fear/14011.jpg'
print("Original image is of fear")
img1 = ef(image1)
pred1 = model.predict(img1)
print(pred1)
pred_label1 = label[pred1.argmax()]
print("The model prediction is ",pred_label1)
plt.imshow(img1.reshape(48,48),cmap='gray')
````
